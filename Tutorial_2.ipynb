```python
#Activity-I: 
#Please report the output of histogram plot. Does it look informative?
not_attack_subset = []
not_attack_subset = data_table[np.where((data_table['label']!=1)& (data_table['port_no']!=3))]
fig =  plt.figure()
plt.hist(not_attack_subset['rx_bytes']-not_attack_subset['tx_kbps'], bins=5)
plt.show()
```

```python
#Activity-II

import matplotlib.pyplot as plt
import numpy as np


# Create a subset where label is not equal to 1 and port number is not equal to 3
not_attack_subset = data_table[np.where((data_table['label'] != 1) & (data_table['port_no'] != 3))]

# Create a figure
fig = plt.figure(figsize=(8, 6))

# Plot the histogram of (rx bytes - tx kbps) color distribution with color changed to red
plt.hist(
    not_attack_subset['rx_bytes'] - not_attack_subset['tx_kbps'],  # Data to plot
    bins=50,  # Number of bins
    color='red', 
    histtype='step',  
    label='no attack'  
)

# Customize y-axis ticks with specific positions and font size
plt.yticks([10, 20, 30, 40, 50], fontsize=28)

# Add label to the x-axis with specified text and font size
plt.xlabel('rx_bytes - tx_kbps', fontsize=28)

# Add label to the y-axis with specified text and font size
plt.ylabel('Number', fontsize=28)

# Add legend with specified location and font size
plt.legend(loc=1, fontsize=18)

# Show the plot
plt.show()
```

```python
#Final tweaks to the histogram

fig = plt.figure(figsize=(8,6))
plt.hist(not_attack_subset['rx_bytes']-not_attack_subset['tx_kbps'], bins=50, color='red',histtype='step', label='no attack')
plt.yticks([10,20,30,40,50], fontsize=28)
plt.show()
```
